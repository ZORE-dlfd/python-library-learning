# Day 1：NumPy ndarray 基础

这一天主要学习 NumPy 中 ndarray 的基本概念和创建方法。

具体内容包括：
- 导入 NumPy，并使用 `np.array()` 创建 0 维、1 维和 2 维数组。
- 查看数组的常用属性：`dtype`、`ndim`、`shape`、`size`。
- 使用 `dtype` 指定数组数据类型。
- 使用 `copy()`、`zeros()`、`ones()`、`empty()`、`full()`、`zeros_like()` 创建数组。
- 使用 `arange()`、`linspace()`、`logspace()` 创建等差和对数间隔数组。
- 使用 `eye()` 和 `diag()` 创建单位矩阵、对角矩阵。
- 练习 NumPy 随机数：`rand()`、`uniform()`、`randint()`、`randn()`。
- 使用 `random.seed()` 固定随机数结果，方便复现实验。
