数据可视化

```python
#折线图
import matplotlib.pyplot as plt
from matplotlib import  rcParams
rcParams['font.sans-serif'] = ['SimHei']


#创建图表，设置大小
pt1 =plt.figure(figsize=(10,5))
month = ['1月','2月','3月 ','4','5','6','7','8','9','10']
sales = ['100','200','300','400','500','600','700','800','900','1000']

plt.plot(month,sales,label='产品A')
#添加标题
plt.title('第一次画',color='green')
plt.xlabel('月份',fontsize=20)
plt.ylabel('销售额',fontsize=20)
plt.xticks(rotation=30,fontsize=20)
plt.yticks(fontsize=20)
plt.grid(True,alpha = 0.4)#网格线 , 透明度
plt.legend(loc='upper left',fontsize=20)


#显示图表
plt.show(pt1)
```

```python
#条形图
pt2 =plt.figure(figsize=(10,5))
subjects = ['语文','math','english']
scores =[50,60,60]
plt.bar(subjects,scores)#横着是barh
plt.ylim(0,100)
plt.show(pt2)
```

```python
#饼图
pt3 =plt.figure(figsize=(10,5))
things = ['xue', 'chi','wan']
times = [3,2,1]
plt.pie(times,labels=things,autopct='%.2f%%',startangle=90,explode=[0,0,0.5])
plt.tight_layout()
```

```python
import random
import matplotlib.pyplot as plt

# 创建画布
pt4 = plt.figure(figsize=(10,8))
scores=[1,2,3,4,5,7]
hours = [1,2,3,4,5,6]
x = []
y = []


for i in range(100):
    tmp = random.uniform(0,10)
    x.append(tmp)

    tmp2 = 2*tmp + random.gauss(0,1)
    y.append(tmp2)

# 绘图：如果要画循环生成的随机散点，改用 plt.scatter(x, y)
plt.scatter(x, y)
plt.plot([0,10],[0,20])#画回归线
plt.show()
```
