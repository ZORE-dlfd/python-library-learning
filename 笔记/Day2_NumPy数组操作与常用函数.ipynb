数据类型


```python
import numpy as np
```

```python
arr = np.array([1,0,1,0],dtype=bool)
print(arr)
```

索引，切片

```python
arr = np.random.randint(1,50,20)
print(arr)
print(arr[5])
print(arr[:])#全部数据
print(arr[2:4])#左开右闭
print(arr[arr>30])
print(arr[(arr>20)&(arr<50)])
print(arr[slice(2,4)])
```

```python
arr = np.random.randint(1,100,(4,8))
print(arr)
print(arr[1,3])
print(arr[1][3])
print(arr[1,:])
print(arr[:][3])#第四行
print(arr[:,3])#第4列
print(arr[2][arr[2]>30])#行数据
```

运算（对应位置相加减）

```python
arr = np.array([1,2,3])
print(arr+3)
```

```python
#广播
a= np.array([1,2,3])#1*3
b = np.array([[1],[2],[3]])#3*1
print(a+b)
```

矩阵运算

```python
a= np.array([1,2,3])#1*3
b = np.array([[1],[2],[3]])#3*1
print(a@b)#矩阵的乘法
```

numpy其它常用函数

```python
print(np.sqrt([1,4,9]))
print(np.exp(1))#e^x
print(np.log(np.exp(1)))#ln
print(np.abs(-1))
print(np.power(a,3))
print(np.round([3.2,4.5,4.51]))
print(np.ceil([3.1]))#向下取整是floor
```

```python
np.isnan([1,2,np.nan,4])
```

```python
arr2 = np.random.randint(1,20,8)
print(arr2)
print(np.sum(arr2))
print(np.mean(arr2))#平均值
print(np.median(arr2))#中位数
print(np.var(arr2))#方差    std是标准差，即sqrt（方差）
print(np.max(arr2),np.argmax(arr2))#后面是索引
print(np.percentile(arr2,20))
print(np.cumsum(arr2))#还有一个cumprod
```

比较函数，基本上返回bool

```python
print(np.greater([3,4,5,6,7],4))#还有less equal
arr3 = np.array([1,2,3,4,5])
print(np.where(arr3>2,arr3,0))
arr4 = np.random.randint(1,100,8)
print(arr4)
print(np.where(arr4>=60,'pass','fail'))
print(np.select([arr4>80,(arr4>60)&(arr4<=80),arr4<60],['1','2','3'],default='4'))
```

排序

```python
np.random.seed(0)
arr5  = np.random.randint(1,100,20)
print(arr5)
arr5.sort()
print(arr5)
print(np.sort(arr5))
print(np.unique(arr5))#去重同时排序
```

拼接

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print(np.concatenate((arr1,arr2)))
```

```python
print(np.split(arr5,4))
print(np.split(arr5,[6,12,18]))
```

```python
print(np.reshape(arr5,[4,5]))
```

axis=0是每列    axis=1是每行
