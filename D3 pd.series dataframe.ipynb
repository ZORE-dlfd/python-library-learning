```python
import pandas as pd
import numpy as np

s = pd.Series([1,2,3,4,5])
print(s)
#自定义索引
s = pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
print(s)
s = pd.Series([1,2,3,4,5],index=['a','b','c','d','e'],name='value')
print(s)
```

```python
s = pd.Series({"a":1,"b":2,"c":3,"d":4,"e":5})
print(s)
s1 = pd.Series(s,index=['a','e'])
print(s1)
print(s.index)
print(s.shape,s.ndim,s.size,s.dtypes)
print(s.loc['a'])
print(s.loc['b'])
print(s.loc['c'])
print(s.iloc[0:3])

```

```python
print(s.at['a'])
print(s.at['b'])
print(s.iat[0])
```

```python
#访问数据
print(s[s<3])#s.head()是前五行   tail（）是后五行，括号里面加n是前后n行
```

例题·

```python
np.random.seed(40)
h = pd.Series(np.random.randint(1,100,24),index=pd.date_range('2020-01-01',periods=24,freq='h'))
day_sales = h.resample('D').sum()
print(day_sales)
mask =(h.index.hour>=8)&(h.index.hour<=22)
b = h[mask]
n_b = h[~mask]
print(b.sum()/n_b.sum())
print(h.nlargest(3).index)


```
